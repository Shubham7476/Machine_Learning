```python
import pandas as pd

# List of important columns
important_columns = ['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES',
                     'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 
                     'CASH_ADVANCE', 'PURCHASES_FREQUENCY', 
                     'ONEOFF_PURCHASES_FREQUENCY', 
                     'PURCHASES_INSTALLMENTS_FREQUENCY', 
                     'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX', 
                     'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS', 
                     'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT']

# Create final_data by selecting important columns and dropping rows with missing values
data = data[important_columns].dropna()

# Print shape of final_data to confirm the number of rows and columns
print("Shape of final data:", data.shape)
```

```python
# Calculate correlation matrix for final_data
corr_matrix = data.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(corr_matrix)
```

```python
# Assuming corr_matrix is your correlation matrix
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Plot correlation matrix as a heatmap with the upper triangular part masked
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", mask=mask)
plt.title("Correlation Matrix (with Upper Triangular Masked)")
plt.show()
```

```python

# Find pairs of columns with correlation greater than 0.5 or less than -0.5
high_corr_pairs = []
checked = set()

# Iterate through all pairs of columns in the correlation matrix
for col1 in corr_matrix.columns:
    for col2 in corr_matrix.columns:
        if col1 != col2:
            # Check if absolute correlation is greater than 0.5 or less than -0.5
            correlation_value = corr_matrix.loc[col1, col2]
            if abs(correlation_value) > 0.5 and (col2, col1) not in checked:
                high_corr_pairs.append((col1, col2))
                checked.add((col1, col2))

# Plot scatter plot matrix for the selected pairs
for pair in high_corr_pairs:
    # Create a scatter plot for the pair of variables
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=data, x=pair[0], y=pair[1])
    plt.title(f'Scatter Plot for {pair[0]} and {pair[1]}')
    plt.xlabel(pair[0])
    plt.ylabel(pair[1])
    plt.show()
```

```python
# Histograms for each column
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 20))

for i, col in enumerate(important_columns):
    row, col_num = divmod(i, 4)
    sns.histplot(data[col], bins=5, kde=False, ax=axes[row, col_num])
    axes[row, col_num].set_title(col)
    
    # percentage
    total = len(data[col])
    for p in axes[row, col_num].patches:
        height = p.get_height()
        axes[row, col_num].text(p.get_x() + p.get_width() / 2.,
                                height + 0.01 * total,
                                '{:.2%}'.format(height / total),
                                ha="center")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the list of columns you want to plot
important_columns = ['BALANCE', 'PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT',
                     'PAYMENTS', 'MINIMUM_PAYMENTS', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX']

# Check if all columns exist in the DataFrame
missing_columns = [col for col in important_columns if col not in data.columns]
if missing_columns:
    print(f"Error: The following columns are missing in the DataFrame: {', '.join(missing_columns)}")
else:
    # Calculate number of rows and columns for subplot grid
    num_cols = 3
    num_features = len(important_columns)
    num_rows = (num_features + num_cols - 1) // num_cols

    # Create subplots for boxplots
    fig_box, axes_box = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 4 * num_rows))

    # Flatten the axes for easy iteration
    axes_box = axes_box.flatten()

    # Plot boxplots for each column in important_columns
    for i, col in enumerate(important_columns):
        sns.boxplot(y=data[col], ax=axes_box[i], linewidth=1.0, color='skyblue', data=data)
        axes_box[i].set_title(f'Boxplot of {col}')
        axes_box[i].set_ylabel(col)  # Set y-label as the column name for clarity

    # Adjust layout
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd

# Assuming df is your DataFrame with relevant columns
# Calculate the correlation matrix
corr_matrix = data.corr()

# Find pairs of columns with high collinearity
high_collinear_pairs = []
threshold = 0.80 # Adjust this threshold as needed

for i in range(len(corr_matrix.columns)):
    for j in range(i + 1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            high_collinear_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j]))

# Display or filter high collinear columns
high_collinear_columns = set()
for col1, col2 in high_collinear_pairs:
    high_collinear_columns.add(col1)
    high_collinear_columns.add(col2)

print("Columns with high collinearity:")
print(high_collinear_columns)
```

```python
rcParams['figure.figsize'] = (7, 5);
wcss= []

for i in range(1,11): 
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=42)
    km.fit(data)
    wcss.append(km.inertia_)
    
plt.plot(range(1,11),wcss, marker='o', linestyle='--')
plt.title('Distortion Score Elbow Method', fontsize =20)
plt.xlabel('K-Value')
plt.ylabel('wcss')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import pandas as pd

# Apply t-SNE to reduce dimensions to 3 for visualization
tsne = TSNE(n_components=3, perplexity=30, learning_rate=0.1, n_iter=2000, random_state=42)
X_tsne = tsne.fit_transform(data)

# Create a DataFrame for the reduced t-SNE components
df_tsne = pd.DataFrame(data=X_tsne, columns=['t-SNE Component 1', 't-SNE Component 2', 't-SNE Component 3'])

# Cluster the data using K-Means with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
cluster_labels = kmeans.fit_predict(data)
df_tsne['Cluster'] = cluster_labels

# Set up 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Define a color palette with 5 distinct colors for clusters
palette = sns.color_palette("tab10", 5)

# Plot each cluster in 3D with different colors
for cluster in df_tsne['Cluster'].unique():
    cluster_data = df_tsne[df_tsne['Cluster'] == cluster]
    ax.scatter(cluster_data['t-SNE Component 1'], 
               cluster_data['t-SNE Component 2'], 
               cluster_data['t-SNE Component 3'], 
               label=f'Cluster {cluster}', 
               color=palette[cluster],
               alpha=0.8)

# Set labels and title
ax.set_xlabel('t-SNE Component 1')
ax.set_ylabel('t-SNE Component 2')
ax.set_zlabel('t-SNE Component 3')
plt.title('Clusters Visualized in 3D by K-means')

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
from sklearn.decomposition import PCA

# Initialize PCA to capture 95% of the variance
pca = PCA(n_components=0.95)

# Fit PCA on the data
pca_result = pca.fit_transform(data)

# Calculate the total explained variance ratio
explained_variance_sum = sum(pca.explained_variance_ratio_)
print('Explained variation summation for all components: {}'.format(explained_variance_sum))
```

```python
pca_result.shape
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

wss = []
K = range(1, 21)  # Range of values for number of clusters (K)

# Iterate over a range of values for K and calculate WSS for each
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans = kmeans.fit(pca_result)
    wss_iter = kmeans.inertia_
    wss.append(wss_iter)
```

```python
import matplotlib.pyplot as plt

def pick_random_color():
    colors = ['red', 'green', 'blue', 'orange', 'purple']
    return np.random.choice(colors)

plt.figure(figsize=(10,6))
plt.plot(K,wss, marker = 'o', c = pick_random_color())
plt.xlabel('K')
plt.ylabel('Within-Cluster-Sum of Squared Errors (WSS)')
plt.xticks(K);
```
